出典
[NB]Paiza_C 問題：C126:宿泊費と交通費より

1 行目に新幹線の片道の料金を表す整数 A、
ホテルの一泊あたりの料金を表す整数 B、
インターンシップの回数を表す整数 Nがこの順で半角スペース区切りで与えられます。
入力例  
200 300 3  

```
ABN=input().split(' ')
A=int(ABN[0]) #新幹線の片道の料金
B=int(ABN[1]) #ホテルの一泊あたりの料金

```

続く N 行のうちの i 行目 (1 ≦ i ≦ N) には、  
i 番目のインターンシップの初日を表す整数 x_i と  
最終日を表す整数 y_i がこの順で半角スペース区切りで与えられます。  
入力例  
1 3  
4 6  
8 10  

```
XY = [] #インターンシップの初日を表す整数 X と最終日を表す整数 Y
for i in range(N):
  XY_list=input().split(' ')
  XY.append(XY_list)
```

入力値を表示する
```
print(A,B,N)
print(XY_list) #XY_list入力値をそれぞれの変数に分解して一時的に入れておくリスト
print(XY)      #XY XY_listを格納する配列
print(XY[1])    #確認用 配列XYの1番地目を表示
print(XY[1][1]) #確認用 配列XYの1番地目の1番地目を表示
```



````python
# 出典
#[NB]Paiza_C 問題：C126:宿泊費と交通費より

# coding: utf-8
# 自分の得意な言語で
# Let's チャレンジ！！
ABN=input().split(' ')
A=int(ABN[0]) #新幹線の片道の料金
B=int(ABN[1]) #ホテルの一泊あたりの料金
N=int(ABN[2]) #インターンシップの回数

XY = [] #インターンシップの初日を表す整数 X と最終日を表す整数 Y
for i in range(N):
  XY_list=input().split(' ')
  XY.append(XY_list)

# print(A,B,N)
# print(XY_list) #XY_list入力値をそれぞれの変数に分解して一時的に入れておくリスト
# print(XY)      #XY XY_listを格納する配列
# print(XY[1])    #確認用 配列XYの1番地目を表示
# print(XY[1][1]) #確認用 配列XYの1番地目の1番地目を表示

# StayDays 自費負担ホテル宿泊日数(インターンシップの間隔日数に一致する)
#インターンシップ１の終了日と２の開始日の間隔がホテル宿泊日数になる
StayDays = []
for j in range(N-1):
  StayDays.append(int(XY[j+1][0]) - int(XY[j][1]))

# StayFee(宿泊金額)とFare(往復運賃)を比較して安いほうを選ぶ
Result = 0
for k in range(len(StayDays)):
  if StayDays[k]*B > A*2:
    Result = Result + A*2
  else:
    Result = Result + StayDays[k]*B
Result = Result + A*2 #初日と最終日の交通費を加算

print(Result)
````
